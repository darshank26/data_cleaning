```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"/Users/darshan/Documents/ML/Data/loan.csv")
dataset.head(10)
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset['Loan_Amount_Term'].fillna(dataset['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Loan_Amount_Term"])
plt.show()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
ft.fit(dataset[['CoapplicantIncome']])
dataset['CoapplicantIncome_new'] = ft.transform(dataset[['CoapplicantIncome']])
dataset
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_new"])
plt.show()
```
