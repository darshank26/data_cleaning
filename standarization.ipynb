```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"/Users/darshan/Documents/ML/Data/loan.csv")
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset['LoanAmount'].fillna(dataset['LoanAmount'].mean(),inplace=True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["LoanAmount"])
plt.show()
```

```python
sns.boxplot(x="LoanAmount",data=dataset)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[['LoanAmount']])
dataset['LoanAmount_ss'] = pd.DataFrame(ss.transform(dataset[['LoanAmount']]))
dataset
```

```python
sns.distplot(dataset["LoanAmount_ss"])
plt.show()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["LoanAmount_ss"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["LoanAmount"])
plt.show()
```
