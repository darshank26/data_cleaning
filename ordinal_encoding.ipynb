```python
import pandas as pd
```

```python
df = pd.DataFrame({"sizes":["S","M","L","XL","M","S","L","M","S","XL","S","L","XXL","XL","S","L"]})
df.head(3)
```

```python
ord_data = [["S","M","L","XL","XXL"]]
ord_data
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# from Sklearn

ord = OrdinalEncoder(categories=ord_data)
ord.fit(df[["sizes"]])
df["en_sizes"]=ord.transform(df[["sizes"]])
df
```

```python
# From map 

ord_data_map = {"S" : 2 ,"M": 4 ,"L" : 8 ,"XL" : 10 ,"XXL" : 12 }
ord_data_map
```

```python
df["en_sizes_map"] = df["sizes"].map(ord_data_map)
df
```
